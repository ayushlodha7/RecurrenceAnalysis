## Implementing Rossler's Attractor for dt =0.05 and Step_Count = 2000

```python
dt = 0.09
step_count = 1000

xs=np.empty((step_count + 1,))
ys=np.empty((step_count + 1,))
zs=np.empty((step_count + 1,))
# The initial conditions.
xs[0], ys[0], zs[0] = (1.0, 1.0, 1.0)
# Iterate
for i in range(step_count):
    x_dot, y_dot, z_dot = Rossler(xs[i], ys[i], zs[i])
    xs[i+1] = xs[i] + (x_dot*dt)
    ys[i+1] = ys[i] + (y_dot*dt)
    zs[i+1] = zs[i] + (z_dot*dt)
fig=plt.figure(figsize=(8,12))
ax=Axes3D(fig)
ax.plot(xs, ys, zs, lw=0.5)
ax.set_xlabel("x", fontsize=15)
ax.set_ylabel("y", fontsize=15)
ax.set_zlabel("z", fontsize=15)
plt.tick_params(labelsize=15)
ax.set_title("Rossler Attractor", fontsize=15)
plt.show()
```

## Plotting the Symbollic Recurrence Plot of Rossler's Attractor

```python
r = np.sqrt(xs**2+ys**2+zs**2)
matrix_rossler = plot_matrix(symbols_for_all_time_series(r,6))
matrix_rossler[matrix_rossler ==0] = np.nan
plt.figure(figsize=(8,12))
plt.imshow(matrix_rossler,cmap="cool",interpolation="none",vmin=0)
plt.title("SRP of Rossler Attractor")
plt.xlabel("Time")
plt.ylabel("Time")
plt.show() 
```

## Symbollic Quantification of Rossler's Attractor.

```python
print("Quantification of Symbollic Recurrence Plot of Rossler Attractor")
Quantification_of_Recurrence_Plot_SRP(matrix_rossler)
```

# Implementing Lorenz Attractor for dt = 0.05 and Step_Count = 1000

```python
# Lorenz parameters and initial conditions
sigma, beta, rho = 3, 1,26.5 
x0, y0, z0 = 0.25, 0.1, 0.1
# Maximum time point and total number of time points
tmax, n = 50, 1000
# Integrate the Lorenz equations on the time grid t.
t = np.linspace(0, tmax, n)
f = odeint(Lorenz, (x0, y0, z0), t, args=(sigma, beta, rho))
x, y, z = f.T
# Plot the Lorenz attractor using a Matplotlib 3D projection.
fig=plt.figure(figsize=(8,12))
ax = Axes3D(fig)
ax.plot(x, y, z, "b-", lw=0.5)
ax.set_xlabel("x", fontsize=15)
ax.set_ylabel("y", fontsize=15)
ax.set_zlabel("z", fontsize=15)
plt.tick_params(labelsize=15)
ax.set_title("Lorenz Attractor", fontsize=15)
plt.show()
```

## Plotting the Recurrence Plot of Lorenz Attractor

```python
r = np.sqrt(x**2+y**2+z**2)
matrix_lorenz = plot_matrix(symbols_for_all_time_series(r,6))
matrix_lorenz[matrix_lorenz ==0] = np.nan
plt.figure(figsize=(8,12))
plt.imshow(matrix_lorenz,cmap="cool",interpolation="none",vmin=0)
plt.title("SRP of Lorenz Attractor")
plt.xlabel("Time")
plt.ylabel("Time")
plt.show() 
```

## Symbollic Quantification Analysis of Lorenz Attractor

```python
print("Quantification of Symbollic Recurrence Plot of Lorenz Attractor")
Quantification_of_Recurrence_Plot_SRP(matrix_lorenz)
```
